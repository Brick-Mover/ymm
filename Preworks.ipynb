```python
# number of users
user_list = np.unique(data['user_id'])
len(user_list)
```

```python
# number of cargoes
len(np.unique(data['id']))
```

```python
data = data.sort_values(by = ['user_id','id'])
print (data.head())
```

```python
# number of cargoes that a user may be interested in 
interest_number_list ={}
for id in user_list:
    tmp_data = data[data['user_id'] == id]
    t = (len([x for x in tmp_data['label'] if x == 1]), len([x for x in tmp_data['label'] if x == 2]))
    r = float(t[0] + t[1])/float(len(tmp_data))
    interest_number_list[id] = (t,r)
```
